```python
from tkinter import *
```

```python
from PIL import Image, ImageTk
import PIL.Image
```

```python
from tkinter.messagebox import *
```

```python
import math
```

```python
root=Tk()
root.title('MY CALCULATOR')
root.geometry('310x400')

font='lucida 22 bold underline'
#important functions
def clear():
    ex=textfield.get()
    ex=ex[0:len(ex)-1]
    textfield.delete(0,END)
    textfield.insert(0,ex)
def all_clear():
    textfield.delete(0,END)
    
def click_button(event):
    b=event.widget
    text=b['text']#buttens ke upar ka text print hoga
    if text=='x':
        textfield.insert(END,'*')
        return
    
    if text=='=':
        try:
            ex=textfield.get()
            ans=eval(ex)
            textfield.delete(0,END)
            textfield.insert(0,ans)
        except Exception as e:
            #print('Error..',e)
            showerror('Error',e)
        return
    if text==',':
    
        pass
    
    textfield.insert(END,text)
    
    

#picture label
image=PIL.Image.open('cal1.jpg')
image=image.resize((40,30))
photo=ImageTk.PhotoImage(image)
label=Label(image=photo)
label.pack(side=TOP,anchor='n')

#heading label
lbl=Label(root,text='My Calculator',font=font)
lbl.pack(side=TOP,fill=X)

#textfield
textfield=Entry(root,font='lucida 22 bold',justify=CENTER)
textfield.pack(side=TOP,padx=5,pady=5,fill=X)

#Creating frame
f1=Frame(root)
f1.pack(side=TOP)
#adding buttens to frame
num=1
for i in range(3):
    for j in range(3):
        btn=Button(f1,text=str(num),font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
        btn.grid(row=i,column=j,padx=2,pady=2)
        num+=1
        btn.bind('<Button-1>',click_button)#left click of mouse

zerobtn=Button(f1,text=0,font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
zerobtn.grid(row=3,column=0,padx=2,pady=2)

dotbtn=Button(f1,text='.',font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
dotbtn.grid(row=3,column=1,padx=2,pady=2)

eqlbtn=Button(f1,text='=',font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
eqlbtn.grid(row=3,column=2,padx=2,pady=2)


plusbtn=Button(f1,text='+',font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
plusbtn.grid(row=0,column=4,padx=2,pady=2)

minusbtn=Button(f1,text='-',font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
minusbtn.grid(row=1,column=4,padx=2,pady=2)


mulbtn=Button(f1,text='x',font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
mulbtn.grid(row=2,column=4,padx=2,pady=2)

divbtn=Button(f1,text='/',font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
divbtn.grid(row=3,column=4,padx=2,pady=2)


backbtn=Button(f1,text='<--',font='lucida 18 bold',command=clear,relief='sunken',width=4,activeforeground='white',activebackground='green')
backbtn.grid(row=4,column=0,padx=2,pady=2)


commabtn=Button(f1,text=',',font='lucida 18 bold',relief='sunken',width=4,activeforeground='white',activebackground='green')
commabtn.grid(row=4,column=1,padx=2,pady=2)

allclrbtn=Button(f1,text='AC',font='lucida 18 bold',bg='grey',relief='sunken',width=4,command=all_clear,activeforeground='white',activebackground='green')
allclrbtn.grid(row=4,column=2,padx=2,pady=2)

exitbtn=Button(f1,text='Exit',font='lucida 18 bold',command=root.destroy,bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
exitbtn.grid(row=4,column=4,padx=2,pady=2)

#binding all buttens
zerobtn.bind('<Button-1>',click_button)
dotbtn.bind('<Button-1>',click_button)
eqlbtn.bind('<Button-1>',click_button)
plusbtn.bind('<Button-1>',click_button)
minusbtn.bind('<Button-1>',click_button)
mulbtn.bind('<Button-1>',click_button)
divbtn.bind('<Button-1>',click_button)
commabtn.bind('<Button-1>',click_button)

#scientific calculator functions
normalcalc=True
def cal_sc(event):
    btn=event.widget
    text=btn['text']
    ex=textfield.get()
    ans=''
    if text=='Deg':
        ans= str(math.degrees(float(ex)))
        
    elif text=='Rad':
        ans= str(math.radians(float(ex)))
    elif text=='x!':
        ans= str(math.factorial(int(ex)))
    elif text=='sinθ':
        ans=math.sin(math.radians(int(ex)))
        pass
    elif text=='cosθ':
        ans=math.cos(math.radians(int(ex)))
    elif text=='tanθ':
        ans=math.tan(math.radians(int(ex)))
    elif text=='√':
        ans=math.sqrt(int(ex))
    elif text=='^':
        base,power=ex.split(',')
        ans=math.pow(int(base),int(power))
                 
    textfield.delete(0,END)   
    textfield.insert(0,ans)

def sc_click():
    global normalcalc
    if normalcalc:
        #sc..
        f1.pack_forget()
        root.geometry('310x520')
        #add sc frame
        sc_f.pack(side=TOP)
        f1.pack(side=TOP)
        normalcalc=False
    else:
        sc_f.pack_forget()
        root.geometry('310x400')
        normalcalc=True
        
    
#creating menu bar
menubar=Menu(root)
mode=Menu(menubar,tearoff=0)
mode.add_checkbutton(label='Scientific Calculator',command=sc_click)
menubar.add_cascade(label='Mode',menu=mode)
root.config(menu=menubar)

#creating scientific frame
sc_f=Frame(root)

#creating buttons

sqrtbtn=Button(sc_f,text='√',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
sqrtbtn.grid(row=0,column=0,padx=2,pady=2)

powbtn=Button(sc_f,text='^',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
powbtn.grid(row=0,column=1,padx=2,pady=2)


factbtn=Button(sc_f,text='x!',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
factbtn.grid(row=0,column=2,padx=2,pady=2)

radbtn=Button(sc_f,text='Rad',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
radbtn.grid(row=0,column=3,padx=2,pady=2)


degbtn=Button(sc_f,text='Deg',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
degbtn.grid(row=1,column=3,padx=2,pady=2)

sinbtn=Button(sc_f,text='sinθ',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
sinbtn.grid(row=1,column=0,padx=2,pady=2)


cosbtn=Button(sc_f,text='cosθ',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
cosbtn.grid(row=1,column=1,padx=2,pady=2)

tanbtn=Button(sc_f,text='tanθ',font='lucida 18 bold',bg='grey',relief='sunken',width=4,activeforeground='white',activebackground='green')
tanbtn.grid(row=1,column=2,padx=2,pady=2)
#binding all buttons
sqrtbtn.bind('<Button-1>',cal_sc)
powbtn.bind('<Button-1>',cal_sc)
factbtn.bind('<Button-1>',cal_sc)
radbtn.bind('<Button-1>',cal_sc)
degbtn.bind('<Button-1>',cal_sc)
sinbtn.bind('<Button-1>',cal_sc)
cosbtn.bind('<Button-1>',cal_sc)
tanbtn.bind('<Button-1>',cal_sc)

root.mainloop()
```
