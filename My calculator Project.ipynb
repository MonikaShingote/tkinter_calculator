```python
from tkinter import *
```

```python
from PIL import Image, ImageTk
import PIL.Image
```

```python
from tkinter.messagebox import *
```

```python
root=Tk()
root.title('MY CALCULATOR')
root.geometry('280x400')

font='lucida 22 bold underline'
#important functions
def clear():
    ex=textfield.get()
    ex=ex[0:len(ex)-1]
    textfield.delete(0,END)
    textfield.insert(0,ex)
def all_clear():
    textfield.delete(0,END)
    
def click_button(event):
    b=event.widget
    text=b['text']#buttens ke upar ka text print hoga
    if text=='x':
        textfield.insert(END,'*')
        return
    
    if text=='=':
        try:
            ex=textfield.get()
            ans=eval(ex)
            textfield.delete(0,END)
            textfield.insert(0,ans)
        except Exception as e:
            #print('Error..',e)
            showerror('Error',e)
        return;
    
    textfield.insert(END,text)
    
    

#picture label
image=PIL.Image.open('cal1.jpg')
image=image.resize((40,30))
photo=ImageTk.PhotoImage(image)
label=Label(image=photo)
label.pack(side=TOP,anchor='n')

#heading label
lbl=Label(root,text='My Calculator',font=font)
lbl.pack(side=TOP,fill=X)

#textfield
textfield=Entry(root,font='lucida 22 bold',justify=CENTER)
textfield.pack(side=TOP,padx=5,fill=X)

#Creating frame
f1=Frame(root)
f1.pack(side=TOP)
#adding buttens to frame
num=1
for i in range(3):
    for j in range(3):
        btn=Button(f1,text=str(num),font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
        btn.grid(row=i,column=j,padx=2,pady=2)
        num+=1
        btn.bind('<Button-1>',click_button)#left click of mouse

zerobtn=Button(f1,text=0,font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
zerobtn.grid(row=3,column=0,padx=2,pady=2)

dotbtn=Button(f1,text='.',font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
dotbtn.grid(row=3,column=1,padx=2,pady=2)

eqlbtn=Button(f1,text='=',font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
eqlbtn.grid(row=3,column=2,padx=2,pady=2)


plusbtn=Button(f1,text='+',font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
plusbtn.grid(row=0,column=4,padx=2,pady=2)

minusbtn=Button(f1,text='-',font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
minusbtn.grid(row=1,column=4,padx=2,pady=2)


mulbtn=Button(f1,text='x',font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
mulbtn.grid(row=2,column=4,padx=2,pady=2)

divbtn=Button(f1,text='/',font='lucida 18 bold',relief='sunken',width=3,activeforeground='white',activebackground='green')
divbtn.grid(row=3,column=4,padx=2,pady=2)


backbtn=Button(f1,text='<--',font='lucida 18 bold',command=clear,relief='sunken',width=7,activeforeground='white',activebackground='green')
backbtn.grid(row=4,column=0,padx=2,pady=2,columnspan=2)

allclrbtn=Button(f1,text='AC',font='lucida 18 bold',bg='grey',relief='sunken',width=3,command=all_clear,activeforeground='white',activebackground='green')
allclrbtn.grid(row=4,column=2,padx=2,pady=2)

exitbtn=Button(f1,text='Exit',font='lucida 18 bold',command=root.destroy,bg='grey',relief='sunken',width=3,activeforeground='white',activebackground='green')
exitbtn.grid(row=4,column=4,padx=2,pady=2)

#binding all buttens
zerobtn.bind('<Button-1>',click_button)
dotbtn.bind('<Button-1>',click_button)
eqlbtn.bind('<Button-1>',click_button)
plusbtn.bind('<Button-1>',click_button)
minusbtn.bind('<Button-1>',click_button)
mulbtn.bind('<Button-1>',click_button)
divbtn.bind('<Button-1>',click_button)

root.mainloop()
```
